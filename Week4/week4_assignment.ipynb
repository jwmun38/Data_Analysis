```python
sales_per_store = X.groupby(['date', 'store_nbr'])['sales'].sum().reset_index()
sales_pivot = sales_per_store.pivot(index='date', columns='store_nbr', values='sales')
sales_per_store
```

```python
sales_pivot.plot(figsize=(15, 8), linewidth=1)
plt.title('Total Sales for Each Store Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.legend(title='Store Number')
plt.xticks(rotation=45)
plt.tight_layout()
```
