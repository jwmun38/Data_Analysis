```python
import matplotlib.pyplot as plt

x = trueskill['game_count']
y = trueskill['uncertainty']

plt.scatter(x, y,marker='o')
plt.ylabel('Uncertainty (σ)',fontsize=10)
plt.xlabel('Game Count',fontsize=10)
plt.title('Decrease in Uncertainty Over Game Count in TrueSkill Rating System',pad=13,fontsize=11)
plt.grid(True)
plt.show()
```

```python
# Plot line graph
plt.figure(figsize=(10, 6))
plt.plot(trueskill['game_count'], trueskill['uncertainty'], marker='o')

plt.xlabel('Uncertainty (σ)')
plt.ylabel('Game Count')
plt.title('Game Count by Uncertainty Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```
