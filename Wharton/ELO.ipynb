```python
import pandas as pd
import numpy as np

def update_elo_by_game_id(df, ratings=None, k=32, default_rating=1500):
    """
    각 game_id별로 팀 매칭 후 Elo 점수를 업데이트하고 df에 elo_before, elo_after 컬럼 추가

    df: 경기 데이터 (각 game_id별로 두 팀이 존재)
    ratings: {팀명: 현재 Elo 점수} 딕셔너리 (없으면 자동 생성)
    k: K-팩터 (점수 변화량 조절)
    default_rating: 새로운 팀의 기본 Elo 점수 (기본값 1500)
    """
    if ratings is None:
        ratings = {}  # Elo 점수 저장 딕셔너리

    new_ratings = ratings.copy()
    df = df.copy()  # 원본 데이터프레임 보호

    # elo_before 컬럼 추가 (경기 전 Elo)
    df["elo_before"] = df["team"].map(lambda team: new_ratings.get(team, default_rating))

    # 경기별 업데이트 적용
    for game_id, game_data in df.groupby("game_id"):
        if len(game_data) != 2:
            continue  # 오류 방지 (경기가 2팀이 아닌 경우)

        team1, team2 = game_data["team"].values
        score1, score2 = game_data["team_score"].values

        # Elo 점수가 없으면 기본값 1500 설정
        if team1 not in new_ratings:
            new_ratings[team1] = default_rating
        if team2 not in new_ratings:
            new_ratings[team2] = default_rating

        rating1, rating2 = new_ratings[team1], new_ratings[team2]

        # 승리/패배 결정
        if score1 > score2:
            result1, result2 = 1, 0
        elif score1 < score2:
            result1, result2 = 0, 1
        else:
            result1, result2 = 0.5, 0.5  # 무승부

        # 예상 승률 계산 (Elo 공식)
        expected1 = 1 / (1 + 10 ** ((rating2 - rating1) / 400))
        expected2 = 1 - expected1

        # Elo 점수 업데이트
        new_ratings[team1] = rating1 + k * (result1 - expected1)
        new_ratings[team2] = rating2 + k * (result2 - expected2)

    # elo_after 컬럼 추가 (경기 후 Elo)
    df["elo_after"] = df["team"].map(lambda team: new_ratings.get(team, default_rating))

    return df

elo_updated_df = update_elo_by_game_id(df)

elo_updated_df
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='elo_after', y='team', data=elo_filtered_df.head(15))  # df가 데이터프레임이라고 가정
plt.xticks(rotation=90)  # 팀명이 길 경우 가로로 돌려서 보이게
plt.xlim(1600, 1900)
plt.title('Win Percentage')
plt.show()
```

```python
elo_updated_df["win"] = (elo_updated_df["elo_after"] > elo_updated_df["elo_before"]).astype(int)

# Elo 점수와 승패의 상관관계 계산
correlation_matrix = elo_updated_df[["elo_before", "elo_after", "win"]].corr()

# 상관관계 히트맵 그리기
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation between Elo Score and Win")
plt.show()

# 상관관계 출력
print(correlation_matrix)
```
