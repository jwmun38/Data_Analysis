```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder
from matplotlib.colors import ListedColormap
```

```python
df = pd.read_csv("./titanic.csv")
```

```python
# 데이터 전처리 
data = df[['survived', 'age', 'pclass', 'sex']].copy()

# 결측치 처리: 나이의 결측치를 중앙값으로 채움
data['age'] = data['age'].fillna(data['age'].median())

# 'sex' 열 인코딩: 남성=1, 여성=0
data.loc[data['sex'] == 'male', 'sex'] = 1
data.loc[data['sex'] == 'female', 'sex'] = 0

# 특성과 타겟 정의
# 시각화를 위해 2개로
X = data[['age', 'pclass']]
y = data['survived']
# 생존 여부에 따라 색상 지정
colors = ['red' if survived == 0 else 'blue' for survived in y]

# 산점도 그리기
plt.figure(figsize=(8, 6))
plt.scatter(X['age'], X['pclass'], c=colors, alpha=0.6, edgecolors='k')
plt.xlabel('Age')
plt.ylabel('Pclass')
plt.title('Survival Visualization (Red: Not Survived, Blue: Survived)')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

k = 2
kmeans = KMeans(n_clusters=2, n_init=10, random_state=42)
y_pred = kmeans.fit_predict(X)
```

```python
# 생존 여부에 따라 색상 지정
colors = ['red' if y == 0 else 'blue' for y in y_pred]

# 산점도 그리기
plt.figure(figsize=(8, 6))
plt.scatter(X['age'], X['pclass'], c=colors, alpha=0.6, edgecolors='k')
plt.xlabel('Age')
plt.ylabel('Pclass')
plt.title('Survival Visualization (Red: Not Survived, Blue: Survived)')
plt.grid(True)
plt.show()
```
